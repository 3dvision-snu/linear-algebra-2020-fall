```python
import numpy as np
import requests
import matplotlib.pyplot as plt

from PIL import Image, ImageFilter
from io import BytesIO
%matplotlib inline
```

```python
# download image and show
url1 = 'https://imgur.com/sQlc7oZ.jpg'
url2 = 'https://imgur.com/5K2oALo.jpg'
response1 = requests.get(url1)
response2 = requests.get(url2)
img1 = Image.open(BytesIO(response1.content)).convert('L')
img2 = Image.open(BytesIO(response2.content)).convert('L')
fig, ax = plt.subplots(1,2)
ax[0].imshow(np.asarray(img1), cmap='gray')
ax[1].imshow(np.asarray(img2), cmap='gray')
```

```python
# blur the image
blurry_img1 = img1.filter(ImageFilter.GaussianBlur(4))
blurry_img2 = img2.filter(ImageFilter.GaussianBlur(8))
fig, ax = plt.subplots(1,2)
ax[0].imshow(np.asarray(blurry_img1), cmap='gray')
ax[1].imshow(np.asarray(blurry_img2), cmap='gray')
```
